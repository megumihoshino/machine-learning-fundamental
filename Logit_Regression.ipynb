```python
import pandas as pd

df = pd.read_csv('/content/sample_data/Social_Network_Ads.csv')
df.head()
```

```python
df.info()
```

```python
#menghilangkan 'user ID' krn merup atribut tdk penting utk dipelajari oleh model
#hilangkan dg fungsi drop dan panggil fungsi get_dummies() utk proses One-Hot Encoding

data = df.drop(columns=['User ID'])

#menjlnkan proses one-hot encoding dg get dummies
data = pd.get_dummies(data)
data
```

```python
#memisahkan atribut dan label
predictions = ['Age', 'EstimatedSalary', 'Gender_Female', 'Gender_Male']
X = data[predictions]
y = data['Purchased']

#melakukan normalisasi thd data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
scaled_data = scaler.transform(X)
scaled_data = pd.DataFrame(scaled_data, columns=X.columns)
scaled_data.head()

#membagi data mjd train dan tes menggunakan splittt
from sklearn.model_selection import train_test_split

#bagi data mjd train dan test utk setiap atribut dan label
X_train, X_test, y_train, y_test = train_test_split(scaled_data, y, test_size = 0.2, random_state =1 )

#membua model logit regression
from sklearn import linear_model

#train model dg fungsi fit
model = linear_model.LogisticRegression()
model.fit(X_train, y_train)
#uji akurasi model
model.score(X_test, y_test)
```
